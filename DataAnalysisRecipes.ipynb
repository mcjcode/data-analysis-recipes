```python
def calc_poly(coef,xs) :
    deg = coef.size-1    
    ys = sum([coef[ii]*xs**ii for ii in xrange(deg+1)],0)
    return ys

def regress(df,deg=1) :
    N = len(df.x)
    xx= np.array(df.x).reshape(N,1)
    A = np.concatenate([xx**i for i in range(deg+1)],1)
    C = np.diag(df.sy*df.sy)
    Y = np.array(df.y).reshape(N,1)
    
    At = A.transpose()
    Cinv = np.linalg.inv(C)
    param_covar=np.linalg.inv(At.dot(Cinv).dot(A))

    X = np.linalg.inv(At.dot(Cinv).dot(A)).dot(At.dot(Cinv).dot(Y))
    
    return X, param_covar

def do_plot(df, X, param_covar, deg=1) :
        
    plt.errorbar(df.x,df.y,yerr=df.sy,fmt='ok')
    
    xs = np.linspace(0,300,50)
    ys = calc_poly(X,xs)
    plt.plot(xs,ys,'k')
    
    ch = np.linalg.cholesky(param_covar)
    for _ in xrange(1000) :
        coef = ch.dot(np.random.normal(size=(deg+1,1))) + X
        ys = calc_poly(coef,xs)
        plt.plot(xs,ys,'gray',alpha=0.02)
     
    plt.xlim([0,300])
    plt.ylim([0,700])
    
    plt.xticks(range(0,300,50))
    plt.yticks(range(0,700,100))
    
    grid(True)

    eqn = '$y ='
    eqn += '+'.join(['%.3f(\pm%.3f)x^%d' % (X[ii],sqrt(param_covar[ii,ii]),ii) for ii in xrange(deg+1)])
    eqn += '$'
    title(eqn,size=14)
```

## 3. Pruning Outliers

```python
def regression_w_outliers_llikelihood(params,data,model) :
    
    # unpack the params, data, and model
    m,b,qs,Pb,Yb,Vb = params
    ys = data
    sys,xs = model
    
    stderrs1 = (ys - m*xs - b)/sys
    stderrs2 = (ys - Yb)/Vb
    z1 = -0.5*stderrs1**2.0-np.log(sys)
    z2 = -0.5*stderrs2**2.0-np.log(Vb)
    return sum(z1*qs+z2*(1-qs))
```

```python
import scipy
from scipy import special, misc

def phi(mu,sigma,z) :
    return np.exp(-0.5*((z-mu)/sigma)**2)/(np.sqrt(2*np.pi)*sigma)

def lphi(mu,sigma,z) :
    return -0.5*((z-mu)/sigma)**2 - np.log(np.sqrt(2*np.pi)*sigma)

def chooseln(N, k) :
    return special.gammaln(N+1) - special.gammaln(N-k+1) - special.gammaln(k+1)
```

```python
def regression_w_outliers_lprior(params) :
    m,b,qs,Pb,Yb,Vb = params
    N = len(qs)
    k = sum(qs)
    retval = 0.0
    retval += lphi( 0.0,   5.0,m)                  # m normal with mean 0 and stdev 100
    retval += lphi(30.0, 200.0,b)                  # b normal with mean 30.0 and stdev 100.0
    retval += 0.0 if (0<Pb<1) else -inf          # Pb uniform between 0 and 1
    retval += 0.0 if (250<=Yb<=750) else -inf      # mean for bad values normal mean 100 and stdev 200
    retval += -0.1*Vb if Vb>=0 else -inf # the prior for the vol is exponential with mean 10
    retval += chooseln(N,k)+ k*log(1-Pb)+(N-k)*log(Pb)
    return retval
```

```python
def regression_w_outliers_proposal(params) :

    scale = 0.002
    
    m, b, qs, Pb, Yb, Vb = params
    
    m2 = m + np.random.normal(0.0,10.0*scale)
    b2 = b + np.random.normal(0.0,1000.0*scale)
    Pb2 = Pb + scale*(2*np.random.random()-1)
    Yb2 = Yb + np.random.normal(0.0,100.0*scale)
    Vb2 = np.exp(np.log(Vb) + np.random.normal(0.0,10*scale))
    qs2 = qs.copy()
    if np.random.random() < 0.65 :
        jj = np.random.random_integers(0,qs.shape[0]-1)
        qs2[jj] = (1-qs2[jj])
    return [m2, b2, qs2, Pb2, Yb2, Vb2]
```

```python
def metropolis_hastings(params,data,model,proposal_func,llhood_func,lprior_func,N) :

    retval = [None]*N
    acceptances = 0
    for ii in xrange(N) :        
        params2 = proposal_func(params)
        lf1 = llhood_func(params,data,model)+lprior_func(params)
        lf2 = llhood_func(params2,data,model)+lprior_func(params2)
        retval[ii] = params + [lf1]
        if np.log(np.random.random()) < lf2-lf1 :
            params = params2
            acceptances += 1
    print 'Acceptance Ratio %6.4f%%'% (100*float(acceptances)/N)
    return retval
```

```python
def bootstrap(df,deg=1,M=10000) :

    XX = np.zeros((deg+1,deg+1))
    N = len(df.x)
    Xall, XXall = regress(df,deg)
    
    for _ in xrange(M) :
        samples = np.random.random_integers(0,N-1,size=N)
        X, XX = regress(df.iloc[samples],deg)

        XX += (X-Xall)*(X-Xall).transpose()
    return Xall, XXall, XX/float(M)
```

```python
def cov_to_ellipse(cov) :
    u,d,v = np.linalg.svd(cov,)
    if u[0,0] >= 0 and u[1,0] >= 0 :
        angle = arccos(u[0,0])*(180.0/np.pi)
        width = sqrt(d[0])
        height = sqrt(d[1])
    else :
        angle = arccos(u[0,1])*(180.0/np.pi)
        width = sqrt(d[1])
        height = sqrt(d[0])
    return height, width, angle
```

```python
def bivariate_log_likelihood(params,data,model) :
    m,b,X = params
    x, y = data
    [dfsx, dfsy, dfrxy] = model
    
    dfdet = dfsx**2 * dfsy**2 - (dfrxy*dfsx*dfsy)**2.0
    
    """
    Return the logarithm of the probability that we would
    have observed points at the locations df.x and df.y
    when the points are actually the points X[i], Y[i]
    on the line y=mx+b and the
    observations are contaminated by Gaussian noise with
    covariance matrix

    [[df.sx**2,           df.rxy*df.sx*df.sy],
     [df.rxy*df.sx*df.sy            df.sy**2]]
    
    """
    llhood = 0.0
    covinv = np.zeros((2,2))
    for ii in xrange(len(X)) :
        A = dfsx[ii]**2
        B = dfrxy[ii]*dfsx[ii]*dfsy[ii]
        C = B
        D = dfsy[ii]**2
        
        det = dfdet[ii]
        
        covinv[0,0] =  D;  covinv[0,1] = -C
        covinv[1,0] = -B;  covinv[1,1] =  A
        covinv /= det
        
        Y = m*X[ii]+b
        v = np.array([[x[ii]-X[ii]],[y[ii]-Y]]);
        llhood += -0.5*v.transpose().dot(covinv).dot(v)-0.5*np.log(2.0*np.pi*det)
    
    return llhood

def bivariate_log_prior(params) :
    m,b,X = params
    retval = 0.0
    retval += lphi(500,10000,b)
    retval += lphi(0.0,10.0,m)
    retval += sum(lphi(150.0,1000,xx) for xx in X)
    return retval

def bivariate_proposal(params) :
    scale = 0.005
    m,b,X = params
    m2 = np.random.normal(m,10.0*scale)
    b2 = np.random.normal(b,100.0*scale)
    X2 = X + np.random.normal(0,100.0*scale,(len(X),))
    return [m2,b2,X2]
```

```python
class MetropolisHastingsSolver(object) :
    def __init__(self,data,model) :
        self.data = data
        self.model = model
        self.samples = []
        
    def initial_parameters(self) :
        raise Exception('Method not implemented')
        
    def log_likelihood(self, params) :
        raise Exception('Method not implemented')
        
    def log_prior(self, params) :
        raise Exception('Method not implemented')

    def proposal(self, params) :
        raise Exception('Method not implemented')

    def metropolis_hastings(self,N) :
        params = self.initial_parameters()
        self.samples = [None]*N
        acceptances = 0
        for ii in xrange(N) :        
            params2 = self.proposal(params)
            lf1 = self.log_likelihood(params )+self.log_prior(params)
            lf2 = self.log_likelihood(params2)+self.log_prior(params2)
            self.samples[ii] = list(params) + [lf1]
            if np.log(np.random.random()) < lf2-lf1 :
                params = params2
                acceptances += 1
        print 'Acceptance Ratio %6.4f%%'% (100*float(acceptances)/N)
```

```python
class TwoDimensionalUncertainties(MetropolisHastingsSolver) :

    def __init__(self,data,model) :
        super(TwoDimensionalUncertainties,self).__init__(data,model)
        
        self.N = len(data[0])
        [dfsx, dfsy, dfrxy] = self.model
        self.dfdet = dfsx**2 * dfsy**2 - (dfrxy*dfsx*dfsy)**2.0        

        self.covinv = np.zeros((self.N,2,2))
        self.covinv[:,0,0] = dfsy**2
        self.covinv[:,0,1] = -dfrxy*dfsx*dfsy
        self.covinv[:,1,0] = self.covinv[:,0,1]
        self.covinv[:,1,1] = dfsx**2
        
        twopi=2.0*np.pi
        self.sum_log_twopidet = sum(np.log(twopi*det) for det in self.dfdet)
        
    def initial_parameters(self) :
        params = [2.0,0.0,np.linspace(0.0,300.0,self.N)]
        return params
    
    def log_likelihood(self,params) :
        m,b,X = params
        x, y = self.data
        """
        Return the logarithm of the probability that we would
        have observed points at the locations df.x and df.y
        when the points are actually the points X[i], Y[i]
        on the line y=mx+b and the
        observations are contaminated by Gaussian noise with
        covariance matrix

        [[df.sx**2,           df.rxy*df.sx*df.sy],
         [df.rxy*df.sx*df.sy            df.sy**2]]

        """
        llhood = 0.0
        v = np.vstack((x-X,y-(m*X+b)))
        for ii in xrange(self.N) :
            vi = v[:,ii]
            llhood += vi.transpose().dot(self.covinv[ii]).dot(vi)            
        return -llhood - self.sum_log_twopidet

    def log_prior(self,params) :
        m,b,X = params
        retval  = lphi(500,10000,b)
        retval += lphi(0.0,10.0,m)
        retval += sum(lphi(150.0,1000,xx) for xx in X)
        return retval

    def proposal(self,params) :
        scale = 0.005
        m,b,X = params
        m2 = np.random.normal(m,10.0*scale)
        b2 = np.random.normal(b,100.0*scale)
        X2 = X + np.random.normal(0,100.0*scale,(self.N,))
        return [m2,b2,X2]
```

```python
def tb_line(theta,bperp,xs) :
    """
    Given the angle theta (from the positive x-axis)
    of the normal vector to the line and the
    perpendicular distance from the origin to the
    line, calculate the y-values corresponding to the
    passed in x-values xs
    """
    m = tan(theta-np.pi/2.0)
    b = bperp/cos(theta-np.pi/2.0)
    return m*xs + b

figure(figsize=(15,3))
xs = np.linspace(-100,+100,5)
for i in range(1,6) :
    subplot(1,5,i)
    ys = tb_line(theta=float(i-1)*np.pi/12.0+np.pi/2.0,bperp=10.0,xs=xs)
    plot(xs,ys,'k')
    xlim([-100,+100])
    ylim([-100,+100])
    grid(True)
```

```python
class TwoDimensionalUncertaintiesMarginalized(MetropolisHastingsSolver) :

    def __init__(self,data,model) :
        super(TwoDimensionalUncertaintiesMarginalized,self).__init__(data,model)
        
        self.N = len(data[0])
        [dfsx, dfsy, dfrxy] = self.model
        self.dfdet = dfsx**2 * dfsy**2 - (dfrxy*dfsx*dfsy)**2.0        

        self.cov = np.zeros((self.N,2,2))
        self.cov[:,0,0] = dfsx**2
        self.cov[:,0,1] = dfrxy*dfsx*dfsy
        self.cov[:,1,0] = self.cov[:,0,1]
        self.cov[:,1,1] = dfsy**2

        self.scales = np.array([1.0,10.0])
        
        twopi=2.0*np.pi
        self.sum_log_twopidet = sum(np.log(twopi*det) for det in self.dfdet)
        
        self.xy = np.vstack((self.data[0],self.data[1]))
        
    def initial_parameters(self) :
        #
        # start with a slope of 0.0, through the origin.
        #
        return np.array([np.pi/2.0,0.0])
    
    def log_likelihood(self,params) :
        """
        Return the log of the probability that we would have observed points
        at the locations x and y when the line was given by
        theta, bperp
        """
        theta, bperp = params
        v = np.array([cos(theta),sin(theta)])
        vt = v.transpose()        
        llhood = sum((v.dot(self.xy[:,ii])-bperp)**2/(2.0*vt.dot(self.cov[ii]).dot(v))
                     for ii in xrange(self.N))
        return -llhood - self.sum_log_twopidet

    def log_prior(self,params) :
        bperp = params[1]
        retval  = 0.0 if (-300<bperp<300) else -inf
        # all theta are equally likely.  no contribution.
        return retval

    def proposal(self,params) :
        scale = 0.01
        return scale*self.scales*np.random.normal(size=(2,)) + params
```
